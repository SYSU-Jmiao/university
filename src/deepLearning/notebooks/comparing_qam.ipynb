```python
import cPickle
import matplotlib.pyplot as plt
```

```python
data = cPickle.load(open('/tmp/data_dict.dat', 'rb'))
```

```python
for p in data:
    print p
partial_data ={}
```

```python
for p in data:
    if p[0] in ["QAM16","QAM64",'AM-DSB']:
        if p[1] > 0:
            partial_data[p] = data[p]        
```

```python
for p in partial_data:
    print p
```

```python
qam16_sample = partial_data[('QAM16', 8)][50]
qam64_sample = partial_data[('QAM64', 8)][50]
```

```python
def plot_sample(source,label):
    plt.figure(figsize=(20,10)) 
    for idx, val in enumerate(source):
        plt.subplot(4, 3,idx+1)
        plt.plot(val[0])
        plt.plot(val[1])
        plt.ylabel(label)
    
    plt.tight_layout()
    plt.show()
        
```

```python
plot_sample(partial_data[('QAM16', 8)][0:10],'QAM16')
```

```python
plot_sample(partial_data[('QAM64', 8)][0:10],'QAM64')
```

```python
plot_sample(partial_data[('AM-DSB', 8)][0:10],'AM-DSB')
```
