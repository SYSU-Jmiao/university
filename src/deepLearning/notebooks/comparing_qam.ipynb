```python
import cPickle
import matplotlib.pyplot as plt
```

```python
data = cPickle.load(open('/tmp/data_dict.dat', 'rb'))
```

```python
for p in data:
    print p
partial_data ={}
```

```python
for p in data:
    if p[1] == 16:
        partial_data[p] = data[p]        
```

```python
for p in partial_data:
    print p
```

```python
def plot_sample(source,label):
    plt.figure(figsize=(20,10)) 
    for idx, val in enumerate(source):
        plt.subplot(4, 3,idx+1)
        plt.plot(val[0])
        plt.plot(val[1])
        plt.ylabel(label)
    
    plt.tight_layout()
    plt.show()
        
```

```python
for p in partial_data:
    plot_sample(partial_data[p][0:10],p)
```
